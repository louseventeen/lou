```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import numpy as np

# Generate random data
# data = np.random.randn(1000)

# Fonction inutile car la fonction "sum" existe dj pour une liste :
# def somme(liste):
#     _somme = 0
#     for i in liste:
#         _somme = _somme + i
#     return _somme

def average(list1, list2):
  newList = []
  iterator = 0
  for value in list1:
      newList.append(round((value+list2[iterator])/2)) #moyenne avec arrond  l'entier infrieur
      iterator += 1
  return newList

def percentage(liste):
  newList = []
  total = sum(liste)
  iterator = 0
  for value in liste:
      newList.append(round((value/total)*100, 1)) #pourcentage avec arrondi au dixime
      iterator += 1
  return newList

# "Diagramme  barres" du type Diagramme  bandes (Bar Chart) V2
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
# ------ Source de donnes : https://ausonius.huma-num.fr/ausoViz/ (Curseur sur valeur du tableau > Bouton Droit > Export > CSV Export) -------------
countFrequency = [1000,4000,30000,25000]
rangeHorizontalAxis = ['Snateurs', 'quites', 'Citoyens', 'Enfants, Esclaves et non citoyens']
# ------------------------------------------------------------------------------
ax.bar(rangeHorizontalAxis,countFrequency)
plt.xlabel('classe sociale', labelpad=5)
plt.ylabel('nombre de places')
plt.title('Diagramme  bandes : Nombre de places attribues en fonction de la classe sociale')
plt.show()


# Diagramme en secteurs / circulaire (Circle Graph)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
value = [1, 5, 50, 44]
name = [ 'Snateurs', 'quites', 'Citoyens', 'Enfants, esclaves et non citoyens']
ax.pie(value, labels = name, autopct = '%1.1f%%')
plt.title('Diagramme en secteurs : Reprsentation en pourcentage de la rpartition des personnes selon leur classe sociale')
plt.show()
```
